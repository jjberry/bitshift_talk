# A Simple Classifier

Suppose we have two types of objects that we want to automatically distinguish. We can measure two separate things about them, both of which are continuosly-valued. For example, suppose we manufacture nails and we have two machines that are slightly differently callibrated. We can measure the length and diameter of each nail. We construct a training set for our algorithm by measuring a set of nails produced by each machine. When we plot the measurements we get the following:

```python
#randomly create the two groups
import numpy as np
mean1 = np.array([0.33, 0.67])
mean2 = np.array([0.67, 0.33])
class1 = np.random.normal(mean1, 0.15, (50, 2)).clip(0, 1)
class2 = np.random.normal(mean2, 0.15, (50, 2)).clip(0, 1)

#plot
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(class1[:,0], class1[:,1], 'ro')
plt.plot(class2[:,0], class2[:,1], 'bo')
plt.xlabel('length')
plt.ylabel('diameter')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()
```

Now suppose we have a box of nails and we want to know which machine manufactured the individual nails for quality control purposes. We will do this by first learning a classifier from our training data and then using that classifier to predict which machine created the nails.

There are many types of classifiers and many variations of training algorithms. We will focus on what is known as Logistic Regression, which fits a sigmoidal curve to the data that can then be used to decide between two categories. 

The sigmoid gets its name from the S-like shape it uses to distinguish 2 categories. It is defined as:

$S(x) = \frac{1}{1 + e^{-x}}$

```python
x = np.linspace(-5, 5, 100)
y = 1.0 / (1.0 + np.exp(-x))
plt.plot(x, y, 'k-')
plt.show()
```

This shape of this function reflects the boundary that our brains exhibit during categorical perception tasks. When a stimulus is on  one of the far ends of the continuum, we reliably percieve the correct category. But when the stimulus is somewhere near the boundary, we make more mistakes. 

We will construct labels for our training data using $0$ for one machine and $1$ for the other.

```python
#make the labels
labels1 = np.zeros(50)
labels2 = np.ones(50)

#put both groups together for training the model
X_train = np.concatenate([class1, class2], axis=0)
y_train = np.concatenate([labels1, labels2])

#construct the model and train
from sklearn.linear_model import LogisticRegression
model = LogisticRegression().fit(X_train, y_train)
```

Now that we have a trained model, we can visually examine the sigmoid decision surface.

```python
x1 = np.arange(0, 1, 0.01)
x2 = np.arange(0, 1, 0.01)
x1, x2 = np.meshgrid(x1, x2)
X_vis = np.concatenate([x1.reshape((10000, 1)), x2.reshape((10000, 1))], axis=1)
y_prob = model.predict_proba(X_vis)
y_prob = y_prob[:,1].reshape((100,100))

# plot the sigmoid surface
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
fig = plt.figure(figsize=(6,6))
ax = fig.gca(projection='3d')
ax.plot_surface(x1, x2, y_prob, cmap=cm.coolwarm_r)
ax.set_xlabel('length')
ax.set_ylabel('diameter')
ax.view_init(40, -130)
plt.show()
```

The sigmoid surface shows the probability of being in class1 or class2 depending on the measurements. As we get closer to the middle, the probability gets closer to $0.5$, where there is overlap between the two classes. Using this model, we can give a probability of being in class1 given any pair of measurements.

We can also force a decision by saying that any probability above $0.5$ is classified as class2 and below $0.5$ is class1, although we should be aware that nails with mesurements near the middle will often be misclassified.

```python
y_forced = model.predict(X_vis)
y_forced = y_forced.reshape((100, 100))
fig = plt.figure(figsize=(6,6))
ax = fig.gca(projection='3d')
ax.plot_surface(x1, x2, y_forced, cmap=cm.coolwarm_r)
ax.set_xlabel('length')
ax.set_ylabel('diameter')
ax.view_init(40,-130)
plt.show()
```
