# Introduction

## How machine learning works

Most machine learning problems involve training a model with some data and then using the model to predict something about new data. When the training data includes the thing we want to predict, we have a supervised learning problem. Later we will discuss unsupervised problems, where we don't have labels in the training set.

The oldest machine learning model is simple linear regression, where we use one continuously-valued variable to predict another. 

```python
# generate some data
import numpy as np
x = np.random.uniform(20, 50, 100)
y = x + np.random.normal(0, 4, 100)

# display it
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(x, y, 'ko')
plt.ylim(0, 60)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Here we can see that $x$ and $y$ are correlated. As $x$ increases, so does $y$. Since they are correlated, we can use $x$ to predict  $y$. Remember that a line can be described with the equation

$y = b + mx$

where $m$ is the slope and $b$ is the y-intercept. If we know the values of $m$ and $b$, we can estimate the value of $y$ from the value of $x$. Since we can't determine $y$ exactly from $x$ using a simple linear model (there is a small variation in $y$) we add a term to the equation to account for the error $\epsilon$. It is customary to use $\beta_0$ for $b$ and $\beta_1$ for $m$ when discussing linear regression.

$y = \beta_0 + \beta_1 x + \epsilon$

Simple, right? Ok but how do we know what the values of $\beta_0$, $\beta_1$ and $\epsilon$ should be?

## Parameter estimation with Gradient Descent

There are many ways to find what the values of the parameters should be. For simple linear regression, they can be calculated using a technique called Ordinary Least Squares, which is very efficient. However there is a more general method that can be used with many different machine learning models called Gradient Descent.

To understand gradient descent it is helpful to visualize the parameter space with respect to some cost function. The cost function is the function we use to measure how accurate our model is. For our linear regression problem, we want our predictions to be as close to the true values of $y$ as possible. We can capture this idea using the following:

$cost = \frac{1}{m}\sum_{i=1}^{m}(\hat{y}_i - y_i)^2$

which is the mean squared error. Here $\hat{y}_i$ is the model's prediction for $x_i$ and $y_i$ is the true value. $\hat{y}_i$ is calculated using our linear model equation above, without the $\epsilon$:

$\hat{y} = \beta_0 + \beta_1 x$

If we plot the $cost$ against the possible values of $\beta_0$ and $\beta_1$, we can see that there is a region where the cost is lower than anywhere else.

```python
# define the cost function
def get_cost(x, y, b0, b1):
    y_hat = b0 + x*b1
    cost = np.sum(np.square(y_hat - y)) / len(y)
    return cost

# plot the cost
#import statsmodels.api as sm
#X = sm.add_constant(x)
#results = sm.OLS(y, X).fit()
#print results.summary()
b0 = np.linspace(-10, 10, 100)
b1 = np.linspace(0.5, 1.5, 100)
b0, b1 = np.meshgrid(b0, b1)
theta = np.concatenate([b0.reshape((10000, 1)), b1.reshape((10000, 1))], axis=1)
z = []
for i in range(10000):
    z.append(get_cost(x, y, theta[i,0], theta[i,1]))
cost = np.array(z).reshape((100, 100))
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(b0, b1, cost, cmap=cm.coolwarm)
ax.set_xlabel('b0')
ax.set_ylabel('b1')
ax.view_init(40, -130)
plt.show()
```
